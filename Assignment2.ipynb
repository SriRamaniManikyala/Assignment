```python
# Step 3: Visualize necessary columns and categorize them
numeric_columns_symmetric = ['mpg', 'acceleration']
numeric_columns_skewed = ['cylinders', 'displacement', 'horsepower', 'weight', 'model year']
categorical_columns = ['origin', 'car name']
```

```python
# Plot histograms for numeric columns
for col in numeric_columns_symmetric + numeric_columns_skewed:
    plt.figure(figsize=(8, 4))
    plt.hist(train[col], bins=20, edgecolor='k')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Plot bar charts for categorical columns
for col in categorical_columns:
    plt.figure(figsize=(25, 5))
    train[col].value_counts().plot(kind='bar')
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import FunctionTransformer
import numpy as np

# Define a function to apply log transformation to numeric columns
def log_transform(X):
    return np.log1p(X)

# Define column groups
numeric_columns_symmetric = ['mpg', 'acceleration']
numeric_columns_skewed = ['cylinders', 'displacement', 'horsepower', 'weight', 'model year']
categorical_columns = ['origin', 'car name']

# Create transformers for each column group
numeric_symmetric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

numeric_skewed_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('log_transform', FunctionTransformer(log_transform)),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])



# Combine transformers using ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num_symmetric', numeric_symmetric_transformer, numeric_columns_symmetric),
        ('num_skewed', numeric_skewed_transformer, numeric_columns_skewed),
        ('cat', categorical_transformer, categorical_columns)
    ])




# Create the final pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Transform the data using the pipeline
X_train_transformed = pipeline.fit_transform(train)
X_test_transformed = pipeline.transform(test)
```
